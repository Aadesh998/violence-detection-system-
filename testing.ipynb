```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
def preprocess_frame(frame, target_size=(128, 128)):
    frame = cv2.resize(frame, target_size)
    frame = frame.astype('float32') / 255.0  
    frame = np.expand_dims(frame, axis=0)  
    return frame
```

```python
def detect_violence(frame, model):
    detection = False
    confidence = model.predict(frame) 
    if confidence > 0.5:  
        detection = True
    return detection, confidence
```

```python
class_labels = ['Non-Violence', 'Violence']
cap = cv2.VideoCapture(0)
# model = 'modelnew.h5'
model_path = r'c:\Users\Dell\Downloads\modelnew.h5'  # Replace with the actual path to your model
model = load_model(model_path)
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # detection, confidence = detect_violence(frame, model)
    preprocessed_frame = preprocess_frame(frame)
    predictions = model.predict(preprocessed_frame)
    print(predictions[0][0])
    if predictions > 0.5:
        predicted_class = 1
    else:
        predicted_class = 0
    print(predicted_class)
    label = class_labels[predicted_class]

    cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

    cv2.imshow('Window', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
